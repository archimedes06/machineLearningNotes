#ATTRIBUTES#

````python
#ATTRIBUTES
#a simple social network
socialNetwork = ig.Graph([(0,1), (0,2), (2,3), (3,4), (4,2), (2,5), (5,0), (6,3), (5,6)])

#each graph has a sequence of all the vertices (vs) and edges (es) within the graph, which we can use to
#label the vertices and edges
socialNetwork.vs["name"] = ["Alice", "Bob", "Claire", "Dennis", "Esther", "Frank", "George"]
socialNetwork.vs["age"] = [25, 31, 18, 47, 22, 23, 50]
socialNetwork.vs["gender"] = ["f", "m", "f", "m", "f", "m", "m"]
socialNetwork.es["is_formal"] = [False, False, True, True, True, False, True, False, False]

#you can alter individual vertices and edges by indexing with es or vs
print("is edge 0 formal: "+str(socialNetwork.es[0]["is_formal"]))
````

````python
#edge object have four fields
    #source
    #target
    #tuple (source, target)
    #index (the edge id)
    #attributes (dictionary of the attributes and values associated with the edge)

#vertices have two fields
    #index
    #attributes

#the graph object, in its entirety, can have attributes
socialNetwork["date"] = "03-19-2015"
print "socialNetwork[\"date\"]: "+ socialNetwork["date"]

#attributes can be deleted with the key word del
del socialNetwork["date"]
````

#STRUCTURAL ATTRIBUTES

````python
#degree... can be obtained for all the vertices, or specific ones
#degree of all vertices
socialNetwork.degree()

#get the degree of the 6th vertex
socialNetwork.degree(6)
socialNetwork.vs[6].degree()

#get the degree of the 1st and 3rd vertices
socialNetwork.degree([1,3])
socialNetwork.vs[[1,3]].degree()

#igraph can create global properties (or edge/vertex properties that have to be executed in a global calcualtion)
#such as eigenvector centrality (of each node).   Similar to page rank.
nodeCentrality = ig.Graph.evcent(socialNetwork)
#or
nodeCentrality = socialNetwork.evcent()
````

#Querying vertices#

````python
#consider the following expression
socialNetwork.vs.select(_degree = socialNetwork.maxdegree())["name"]
#positional arguments, the ones without a name, like _degree
    #If the positional argument is None, none of the sequence will be returned
    #if its a partially applied function, they the function will be called on each element of the sequence, if the
    #function returns true, then the element will returned as part of the sequence (think filter in scala)

#recall lambda expressions in python are like anonymous functions in scala
oddElements = socialNetwork.vs.select(lambda vertex: vertex.index % 2 == 1)

#if the positional argument is an iterable, list or a generator that returns ints, vertices corresponding to those
#indexes will be returned
socialNetwork.vs.select([1,2,3,4])

#keyword arguments can be used to filter the vertices based on their attributes, where this means, age attribute less
#than, 30
youngElements = socialNetwork.vs.select(age_lt = 30)

#structural properties (to distinguish them from attributes with the same name) are preceeded by underscore, like
greaterDegree = socialNetwork.vs.select(_degree_gt = socialNetwork.maxdegree()-2)["name"]

#to select edges from a specific vertex... for undirected graphs, it only matters that the edge touches the vertex
#for directed graphs, _target, and _source will return different values
socialNetwork.es.select(_source=2)

#_within returns the all the edges that originate and teminate within the set of indicated vertices
interConnection = socialNetwork.es.select(_within=[2,3,4])
#similarly
interConnectionII = socialNetwork.es.select(_within=socialNetwork.vs[2:5])

#_between, returns the edges that span two VertexSeq
men = socialNetwork.vs.select(gender = "m")
women = socialNetwork.vs.select(gender = "f")
socialNetwork.es.select(_between=(men, women))  #all connections between men and women

#find, works similarly to select, but only returns the first match... and will throw an exception if there are
#no matches
claire = socialNetwork.vs.find("Claire")


#turns out that the name attribute is treated specially, so that if a string is passed to a general function, it
#will match on the name parameter.   Note if two vertices have the same name, only one will be returned.
print "degree of Claire's node: "+str(socialNetwork.degree("Claire"))
````

#Generating the Degree matrix (diagonal, with the degree of each node along the diagonal)

````python
degreeM = np.diag(karate.degree())
print "degreeM: \n"+str(degreeM)
````

<font size=6 color="green">Generating the Laplacian L = D-A</font>

````python
laplacianM = degreeM-adjacencyM
print "laplacianM: \n"+str(laplacianM)
````

````python
#Returns Tuple (values, vectors), where vectors is a matrix, and the eigen vectors are the columns
eigenValues, eigenVectors = np.linalg.eig(laplacianM)
print "Dimensions of the eigenVector matrix: "+ str(eigenVectors.shape)
print "Eigen Velues: \n"+str(eigenValues)
````

````python
#indices for eigenValues, sorted largest -> smallest
sortedEigenValueIndexes = np.argsort(eigenValues, axis = 0)
eigenValues = eigenValues[sortedEigenValueIndexes][1::]
eigenVectors = eigenVectors[:,sortedEigenValueIndexes][:, 1::]
print(eigenVectors.shape)
print "Sorted EigenValues: "+str(eigenValues)

````

````python
#seems that floating point rounding errors can be a problem here... checkout eigenValue Number 8.... even though the eigen value is 
#~5, the components of the eigen vector are of order ~10^-15, and the eigen value equation breaks down 
eigenValue_i = 5
print "Eigen value: " + str(np.dot(laplacianM, eigenVectors[:,eigenValue_i])[0] / eigenVectors[0,eigenValue_i])
print np.dot(laplacianM, eigenVectors[:,eigenValue_i])
print eigenVectors[:,eigenValue_i]
print "Declared second eigen value: "+str(eigenValues[eigenValue_i])
````

<font size=8> DBSCAN </font>

<font size=4>
    DBSCAN basically works by considering the density of points around every point.   If the density is high, then the point in question is added to the cluster that accounts for the nearby density.  Thus, there are two parameters that need to be set, what distance should be considered "nearby" (eps) and how many points should consitute high density.   For more information see
    <a href="http://en.wikipedia.org/wiki/DBSCAN">
        DBSCAN
    </a> 
</font>

````python
from sklearn.cluster import DBSCAN
````

````python
euclideanDistance = lambda a,b:  np.linalg.norm(a-b)
````

<font size=4>Determining the optimum distance for the neighbor search</font> 

````python
def kDistance(graphMatrix, k, distanceMetric = euclideanDistance):
    numberSamples, numberDimensions = graphMatrix.shape
    
    #doing this the hard way, calculate the k-nearest neighbors by brute force
    kDistanceList = []
    for refPoint in range(0, numberSamples):
        refSample = graphMatrix[refPoint, :]
        comparisonList = np.delete(graphMatrix, refPoint, axis=0)
        currentDistances = []
        for comparisonPoint in range(0, numberSamples-1):
            comparisonSample = comparisonList[comparisonPoint, :]
            currentDistances.append(distanceMetric(refSample, comparisonSample))
        
        #at this point, currentDistances should have the distance from the referenceSample, to all other samples
        #the k-smallesst elements in currentDistance
        kSmallest = np.sort(currentDistances)[0:k]
        #avergae of the k smallest
        kDistance = np.average(kSmallest)
        kDistanceList.append(kDistance)
    return kDistanceList
````

<font size=5><B><i>Generating the k-distance plot</B></i></font>

````python
k = 4
kDistanceList = np.sort(kDistance(reducedMatrix, k=k))
figure()
plot(kDistanceList)
title("k = "+str(k)+" Distance")
show()
````

<font size=4>This indicates an epsilon value of ~ 0.4 (the location of the knee in the curve).   Using the formula minPts >= D+1 (and here D is the dimensionality of the cluster points (D=4), we get minPts >= 5</font>

````python
#computing the DBSCAN clusters
eps=0.2
minPts = 5
dbScanClusters = DBSCAN(eps=eps, min_samples = minPts).fit(reducedMatrix)
````
